```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 6 }
        )
plot_correlation_map(df)
plt.show()
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics

y=df['price']
print(y)
x=df[['bedrooms','bathrooms','sqft_living','floors','waterfront','view', 'grade', 'sqft_above','sqft_basement','lat','sqft_living15']]
print(x)
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.2, random_state=30)
# We split our date into 80% for training and 20% for testing 
# In order to compare correctly the results of training we will use the same split for all the following examples
#otherwise the results can be different because the split is done randomly
```

```python
x_sqftliv=df['sqft_living'].values[:,np.newaxis]
x_train_sqftliv=x_train['sqft_living'].values[:,np.newaxis]
x_test_sqftliv=x_test['sqft_living'].values[:,np.newaxis]
y_train2=y_train.values
y_test2=y_test.values
model=LinearRegression()   #build linear regression model
model.fit(x_train_sqftliv,y_train2)  #fitting the training data
predicted=model.predict(x_test_sqftliv) #testing our modelâ€™s performance

#Measuring the performance of linear regression using the testing set
print("MSE", mean_squared_error(y_test2,predicted))
print("R squared", metrics.r2_score(y_test2,predicted))

#plotting the linear regression
plt.scatter(x_sqftliv,y,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x_sqftliv,model.predict(x_sqftliv),color="k")
plt.show()
#We are going to do the same thing for the other selected features
```
